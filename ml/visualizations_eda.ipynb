```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

features_path = 'training_set_features.csv'
labels_path = 'training_set_labels.csv'
features_df = pd.read_csv(features_path)
labels_df = pd.read_csv(labels_path)
```

```python
data_df = pd.merge(features_df, labels_df, on='respondent_id')
```

## General Visualizations : Getting an idea

```python
h1n1_vaccine_counts = data_df['h1n1_vaccine'].value_counts()
plt.figure(figsize=(8, 8))
h1n1_vaccine_counts.plot(kind='pie', labels=['Not Vaccinated', 'Vaccinated'],
                         autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'],
                         explode=(0.1, 0), shadow=True)
plt.title('Proportion of Respondents by H1N1 Vaccination Status')
plt.ylabel('')  # Hiding the y-label as it's not necessary for pie charts
plt.show()
```

```python
seasonal_vaccine_counts = data_df['seasonal_vaccine'].value_counts()
plt.figure(figsize=(8, 8))
seasonal_vaccine_counts.plot(kind='pie', labels=['Not Vaccinated', 'Vaccinated'],
                             autopct='%1.1f%%', startangle=140, colors=['#99ff99','#ffcc99'],
                             explode=(0.1, 0), shadow=True)
plt.title('Proportion of Respondents by Seasonal Flu Vaccination Status')
plt.ylabel('')
plt.show()
```

```python
ax = sns.countplot(x='chronic_med_condition', hue='chronic_med_condition', data=data_df,palette="deep", legend=False)

plt.title('Chronic Medical Condition Prevalence')
plt.xlabel('Has Chronic Medical Condition')
plt.ylabel('Number of Respondents')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
insurance_counts = data_df['health_insurance'].value_counts()
plt.figure(figsize=(8, 8))
insurance_counts.plot(kind='pie', labels=['Covered', 'Not Covered'],
                      autopct='%1.1f%%', startangle=140, colors=['#85c1e9', '#f0b27a'],
                      explode=(0.1, 0), shadow=True)
plt.title('Proportion of Respondents by Health Insurance Coverage')
plt.ylabel('')
plt.show()
```

```python
sns.countplot(x='behavioral_face_mask', data=data_df)
plt.title('Usage of Face Masks')
plt.xlabel('Used Face Mask')
plt.ylabel('Number of Respondents')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

## Levels of Concern and Knowledge about H1N1

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(x='h1n1_concern', hue='h1n1_concern', data=data_df, ax=axes[0], palette='viridis',legend=False)
axes[0].set_title('Levels of Concern about H1N1 Flu')
axes[0].set_xlabel('Level of Concern')
axes[0].set_ylabel('Count')
axes[0].set_xticklabels(['Not at all', 'Not very', 'Somewhat', 'Very'])

sns.countplot(x='h1n1_knowledge', hue='h1n1_knowledge', data=data_df, ax=axes[1], palette='rocket', legend=False)
axes[1].set_title('Levels of Knowledge about H1N1 Flu')
axes[1].set_xlabel('Level of Knowledge')
axes[1].set_ylabel('Count')
axes[1].set_xticklabels(['No knowledge', 'A little', 'A lot'])

plt.tight_layout()
plt.show()
```

## Doctor Recommendations on Vaccinations and Its Influence

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(x='doctor_recc_h1n1', hue='doctor_recc_h1n1', data=data_df, ax=ax[0], palette="coolwarm", legend=False)
ax[0].set_title('Doctor Recommendations for H1N1 Vaccine')
ax[0].set_xlabel('Recommendation')
ax[0].set_ylabel('Number of Respondents')
ax[0].set_xticklabels(['No', 'Yes'])

sns.countplot(x='doctor_recc_seasonal', hue='doctor_recc_seasonal', data=data_df, ax=ax[1], palette="coolwarm",legend=False)
ax[1].set_title('Doctor Recommendations for Seasonal Flu Vaccine')
ax[1].set_xlabel('Recommendation')
ax[1].set_ylabel('Number of Respondents')
ax[1].set_xticklabels(['No', 'Yes'])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='doctor_recc_h1n1', hue='h1n1_vaccine', data=data_df, palette="colorblind")
plt.title('Influence of Doctor’s Recommendation on H1N1 Vaccination')
plt.xlabel('Doctor Recommended H1N1 Vaccine')
plt.ylabel('Number of Respondents')
plt.legend(title='H1N1 Vaccine', labels=['Not Vaccinated', 'Vaccinated'])
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='doctor_recc_seasonal', hue='seasonal_vaccine', data=data_df, palette="colorblind")
plt.title('Influence of Doctor’s Recommendation on Seasonal Flu Vaccination')
plt.xlabel('Doctor Recommended Seasonal Flu Vaccine')
plt.ylabel('Number of Respondents')
plt.legend(title='Seasonal Flu Vaccine', labels=['Not Vaccinated', 'Vaccinated'])
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

## Opinions on Vaccine Effectiveness

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# For H1N1 vaccine opinions
sns.countplot(x='opinion_h1n1_vacc_effective', hue='opinion_h1n1_vacc_effective',data=data_df, ax=ax[0], palette="YlGnBu", legend=False)
ax[0].set_title('Opinions on H1N1 Vaccine Effectiveness')
ax[0].set_xlabel('Effectiveness Rating')
ax[0].set_ylabel('Number of Respondents')
# Set custom x-axis labels
ax[0].set_xticklabels(['Not at all effective', 'Not very effective', 'Don\'t know', 'Somewhat effective', 'Very effective'], rotation=45, ha="right")

# For Seasonal vaccine opinions
sns.countplot(x='opinion_seas_vacc_effective', hue='opinion_seas_vacc_effective',data=data_df, ax=ax[1], palette="YlGnBu",legend=False)
ax[1].set_title('Opinions on Seasonal Vaccine Effectiveness')
ax[1].set_xlabel('Effectiveness Rating')
ax[1].set_ylabel('Number of Respondents')
# Set custom x-axis labels
ax[1].set_xticklabels(['Not at all effective', 'Not very effective', 'Don\'t know', 'Somewhat effective', 'Very effective'], rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

## H1N1 Vaccination Dependency on Different Factors

### 1. Age Group

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='age_group', hue='h1n1_vaccine', data=data_df, palette="deep",
              order=['18 - 34 Years', '35 - 44 Years', '45 - 54 Years', '55 - 64 Years', '65+ Years'])
plt.title('H1N1 Vaccination by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Respondents')
plt.legend(title='H1N1 Vaccine', labels=['Not Vaccinated', 'Vaccinated'])
plt.xticks(rotation=45)
plt.show()
```

### 2. Employment Status

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='employment_status', hue='h1n1_vaccine', data=data_df, palette="Set1")
plt.title('H1N1 Vaccination Rates by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Number of Respondents')
plt.legend(title='H1N1 Vaccine', labels=['Not Vaccinated', 'Vaccinated'])
plt.xticks(rotation=45)
plt.show()
```

### 3. Education Level

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='education', hue='h1n1_vaccine', data=data_df, palette="viridis")
plt.title('H1N1 Vaccination Rates by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.legend(title='H1N1 Vaccine', labels=['Not Vaccinated', 'Vaccinated'])
plt.xticks(rotation=45)
plt.show()
```

### 4. Income Level

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='income_poverty', hue='h1n1_vaccine', data=data_df, palette="plasma")
plt.title('H1N1 Vaccination Rates by Income Level')
plt.xlabel('Income Level')
plt.ylabel('Number of Respondents')
plt.legend(title='H1N1 Vaccine', labels=['Not Vaccinated', 'Vaccinated'])
plt.xticks(rotation=45)
plt.show()
```
